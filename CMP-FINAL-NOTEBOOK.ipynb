What would you consider the “best” neighborhood in Pittsburgh? It’s a loaded question, that of course will have many different answers. But I believe the questions answers can start at the same spot. If you were to build your “perfect neighborhood” - What sort of attributes should it have? What attributes should it not have? How does your definition of the perfect neighborhood differ from the societal view of one (if there even is a set and stone consensus of one). I decided to go the environmental route when comparing neighbor hoods. With 3 different metrics within that area. Let's take a look.  

```python
import pandas 
```

```python
treeData = pandas.read_csv('Trees.csv', converters={'id': convertString, 'address_number': convertString})
```

```python
TRneighborhoods = treeData.neighborhood.unique()
numTreesToNeigh = []

for neighborhood in TRneighborhoods:
    neighborhood2trees = [neighborhood , (treeData.neighborhood == neighborhood).sum()]
    numTreesToNeigh.append(neighborhood2trees)

dataFrameTrees2Neigh = pandas.DataFrame(numTreesToNeigh, columns =['Neighborhood', 'Number Trees'])
dataFrameTrees2Neigh
```

^ Here I took in a data set of every single tree located in the Pittsburgh area and the neighborhood it was in (yeah it was insane amount of data), counted the number of trees per neighborhood and displayed them in a table. The more trees = the better in this metric (as in real life).  

```python
column = dataFrameTrees2Neigh['Number Trees']
treeMax = column.argmax()

maxTreeNeighborhood = dataFrameTrees2Neigh.loc[treeMax]

maxTreeNeighborhood
```

^ Squirrel Hill South wins in this metric with 5073 trees.

```python
parksData = pandas.read_csv ('Parks.csv')
```

```python
parksOnlyDF = parksData[parksData['type'] == 'Park']
```

```python
Pneighborhoods = parksOnlyDF.neighborhood.unique()
numParksToNeigh = []

for neighborhood in Pneighborhoods:
    numParks = (parksOnlyDF.neighborhood == neighborhood).sum()
    neighborhood2park = [neighborhood , numParks]
    numParksToNeigh.append(neighborhood2park)

dataFrameParks2Neigh = pandas.DataFrame(numParksToNeigh, columns = ['Neighborhood', 'Number Parks'])
dataFrameParks2Neigh
```

^ I took a similar approach here with a similar metric - more parks = a better neighborhood. 

```python
parks_column = dataFrameParks2Neigh ['Number Parks']
parksMax = parks_column.argmax()

maxParkNeighborhood = dataFrameParks2Neigh.loc[parksMax]
maxParkNeighborhood
```

^ Beechview wins this metric with 6 parks in its neighborhood.

```python
trashData = pandas.read_csv ('Trash.csv')
```

```python
Tneighborhoods = trashData.neighborhood.unique()
numTrashToNeigh = []

for neighborhood in Tneighborhoods: 
    numTrash = (trashData.neighborhood == neighborhood).sum()
    neighborhood2trash = [neighborhood, numTrash]
    numTrashToNeigh.append(neighborhood2trash)

dataFrameTrash2Neigh = pandas.DataFrame(numTrashToNeigh, columns = ['Neighborhood', 'Number Trash'])
dataFrameTrash2Neigh
```

^ And once again with 'smart trashcans.' Smart trashcans are trashcans that have a sensor in them that sends a signal to the Department of Public Works whenever they are full - preventing an overflow of garbage and reducing street litter. 

```python
trash_column = dataFrameTrash2Neigh['Number Trash']
trashMax = trash_column.argmax()

maxTrashNeighborhood = dataFrameTrash2Neigh.loc[trashMax]
maxTrashNeighborhood
```

^ Shadyside wins this category - with 77 smart trashcans in the neighborhood.

```python
def Union(list1,list2,list3): 
    final_list = list(set(list1) | set(list2) | set(list3)) 
    return final_list 
```

```python
finalNeighborHoods = Union(TRneighborhoods, Pneighborhoods, Tneighborhoods)
```

```python
final_scores = []

for nHood in finalNeighborHoods :
    nTrees = dataFrameTrees2Neigh.loc[dataFrameTrees2Neigh.Neighborhood == nHood, 'Number Trees'].values
    nParks = dataFrameParks2Neigh.loc[dataFrameParks2Neigh.Neighborhood == nHood, 'Number Parks'].values
    nTrash = dataFrameTrash2Neigh.loc[dataFrameTrash2Neigh.Neighborhood == nHood, 'Number Trash'].values
    
    if nTrees.size == 0:
        nTrees = 0
    else:
        nTrees = nTrees[0]
        
    if nParks.size == 0:
        nParks = 0
    else:
        nParks = nParks[0]
        
    if nTrash.size == 0:
        nTrash = 0 
    else:
        nTrash = nTrash[0]
        
    final_score = nTrees + nParks + nTrash
    neigh_final_score = [nHood, nTrees, nParks, nTrash, final_score]
    final_scores.append(neigh_final_score)


dataFrameFinalScores = pandas.DataFrame(final_scores, columns = ['Neighborhood', 'Number Trees', 'Number Parks', 'Number Trashcans', 'Final Score'])

pandas.set_option("display.max_rows", None, "display.max_columns", None)
print(dataFrameFinalScores)
```

^ Up above is the final score computed for every neighborhood being evaluated using a score calculator by adding up each score in each category. By using this data we can see that Squirrel Hill South wins the award for best neighborhood in Pittsburgh based on a metric of environmental friendliness. Of course this metric has its flaws. There is too much weight put on the amount of trees in each neighborhood that will affect the final score. As you'll see down below, the neighborhood of "Hays" was immedietly out of contention with only 1 tree registered in their neighborhood - no where near Squirrel Hill South with 5,073 trees. In order to improve this metric - somehow that weight would need to have a less of an impact on the final score.
